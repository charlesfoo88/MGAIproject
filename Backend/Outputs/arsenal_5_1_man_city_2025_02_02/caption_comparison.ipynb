## 7. Key Observations

### Caption Quality Differences:

**RAG Multi-Agent:**
- More descriptive and engaging (e.g., "what a finish!", "brilliant goal!")
- Includes player context ("Ødegaard scores with Havertz' help")
- Expressive language reflecting fan emotion
- Average 12-13 words per caption
- Full names used ("Ethan Nwaneri", "T. Partey")

**Baseline Single-Prompt:**
- More factual and concise
- Simple structure: "[Player] [action] for [team]"
- Less emotional language
- Average 6-7 words per caption (SHORTER than requested 10-20)
- Minimal context or storytelling

### Performance:
- **Baseline:** ~1.3-1.6 seconds, ~750 tokens per perspective
- **RAG Pipeline:** ~3-5 minutes (multi-agent orchestration, evidence retrieval, hallucination checks)

### Word Count Adherence:
- RAG captions: 10-16 words (mostly within 10-20 range)
- Baseline captions: 4-7 words (BELOW requested 10-20 range)

**This suggests the RAG system better follows the word count constraint, possibly due to multi-step verification.**
